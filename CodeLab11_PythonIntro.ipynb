#### 3. Посчитайте количество слов из white_list в файле log.txt

```python
from collections import Counter
import functools

def get_white_words(line):
    parts = line.strip().split(' ')
    words = parts[2:]
    return [w.lower().strip() for w in words if w.lower().strip() in white_list]

all_white_words = functools.reduce(lambda a, b: a + b, map(get_white_words, lines))
counts = Counter(all_white_words)
print(counts)
```
